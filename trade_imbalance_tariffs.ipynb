CNBC also published a page [explaining how the tariffs were calculated](https://www.cnbc.com/2025/04/05/economists-take-issue-with-trumps-tariff-formula-arguing-rate-is-inflated.html)

Note that the calculation is performed by using the formula:

$$\Delta\tau_i = \frac{x_i - m_i}{\epsilon \times \phi \times m_i}$$

where $x_i$ and $m_i$ are exports to, and imports from (respectively) country $i$, while $\epsilon$ and $\phi$ are factors for elasticity of demand and prices.

However, the geniuses used $\epsilon = 4.0$ and $\phi = 0.25$ which when multiplied in the formula yield a value of 1.0, essentially making the formula collapse to:

$$\Delta\tau_i = \frac{x_i - m_i}{m_i}$$

or simply the difference obtained by subtracting imports from exports and dividing the difference by the imports - again - completely unhinged from the real-worldrates.

```python
import pandas as pd
import lxml
import openpyxl
```

```python
url = "https://worldpopulationreview.com/country-rankings/us-trade-deficit-by-country"
```

```python
def get_table_from_website(url, table_index=0, header=0):
    """
    Retrieves a table from a specified website using pandas.

    Args:
        url (str): The URL of the website containing the table.
        table_index (int): The index of the table to retrieve (default is 0 for the first table).
                           Some websites may have multiple tables.
        header (int or list of int, default=0): The row number(s) to use as the column names.
                                                 0-indexed (0 is the first row). If the table
                                                 doesn't have a header or the header spans
                                                 multiple rows, you can specify a list of row
                                                 indices to create a MultiIndex. Use None if
                                                 the table has no header.

    Returns:
        pandas.DataFrame: A DataFrame containing the data from the specified table,
                          or None if an error occurs.
    """
    try:
        # Use pandas.read_html to read all tables from the URL
        tables = pd.read_html(url, header=header)

        # Check if the specified table index exists
        if table_index < len(tables):
            return tables[table_index]
        else:
            print(f"Error: Table with index {table_index} not found on the website.")
            return None

    except ValueError as e:
        print(f"Error reading HTML table from {url}: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None
```

```python
df_trade_data = get_table_from_website(url, table_index=0, header=0)
df_trade_data
```

```python
df_numeric_trade_data.drop(columns=["Unnamed: 0", "US Trade Deficit 2024↓", "US Total Trade 2024", "US Trade Balance 2024"], inplace=True)
```

```python
df_numeric_trade_data["US Imports from 2024"] = df_numeric_trade_data["US Imports from 2024"].map(
    lambda x: x.replace(",", "").replace("$", "").replace(" ", "")
)

df_numeric_trade_data["US Exports to 2024"] = df_numeric_trade_data["US Exports to 2024"].map(
    lambda x: x.replace(",", "").replace("$", "").replace(" ", "")
)

# .replace("B", "000").replace("M", "")
df_numeric_trade_data
```

```python
def parse_number_with_suffix(value):
    if value.endswith('B'):
        return float(value[:-1]) * 1_000_000_000
    elif value.endswith('M'):
        return float(value[:-1]) * 1_000_000
    elif value.endswith('K'):
        return float(value[:-1]) * 1_000
    else:
        return float(value)
```

```python
df_numeric_trade_data["US Imports from 2024"] = df_numeric_trade_data["US Imports from 2024"].map(lambda x: parse_number_with_suffix(x))
df_numeric_trade_data["US Exports to 2024"] = df_numeric_trade_data["US Exports to 2024"].map(lambda x: parse_number_with_suffix(x))
```

```python
def format_currency(number):
    if abs(number) >= 1_000_000_000:
        return f"${number / 1_000_000_000:,.1f}B"
    elif abs(number) >= 1_000_000:
        return f"${number / 1_000_000:,.1f}M"
    elif abs(number) >= 1_000:
        return f"${number / 1_000:,.1f}K"
    else:
        return f"${number:,.1f}"
```

```python
def reciprocal_tariff_formula(imp, exp, epsilon=4.0, phi=0.25):
    """
    Calculate the reciprocal tariff based on imports and exports.

    Args:
        imp (float): Imports value.
        exp (float): Exports value.

    Returns:
        float: Idiotic tariff value as cacluated by the current "administration"
    """
    if exp == 0:
        return float('inf')  # Avoid division by zero
    
    tariff = (exp - imp) / (epsilon * phi * imp)
    if tariff > -0.1:
        tariff = -0.1 # Apply a minimum 10% tariff (incl. even on countries with which the USA has a positive trade balance)
    
    return -round(tariff, 2)
```

```python
df_numeric_trade_data["Nominal Imbalance"] = df_numeric_trade_data.apply(
    lambda row: reciprocal_tariff_formula(row["US Imports from 2024"], row["US Exports to 2024"]), axis=1
)
```

```python
df_numeric_trade_data["Applied Tariff"] = df_numeric_trade_data.apply(
    lambda row: round(row["Nominal Imbalance"]/2, 2) if round(row["Nominal Imbalance"]/2, 2) > 0.1 else 0.1, axis=1
)
```
